```python
import pandas as pd
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,6],"C":[21,22,23,24,25]})
```

```python
pd.merge(var1,var2,on="A")
```

```python
pd.merge(var1,var2,how="left")
```

```python
pd.merge(var1,var2,how="right")
```

```python
pd.merge(var1,var2,how="outer",indicator=True)
```

## Concat

```python
l1 = pd.Series([1,2,3,4,5])
l2 = pd.Series([11,12,13,14,15])
pd.concat([l1,l2])
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
d2 = pd.DataFrame({"A":[1,2,3,4,6],"C":[21,22,23,24,25]})
```

```python
pd.concat([d1,d2])
```

```python
pd.concat([d1,d2],axis=1)
```
